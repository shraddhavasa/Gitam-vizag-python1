```python
import pandas as pd
filepath='datatfiles/AppleStore.csv'
def readcsv(filepath):
    return pd.read_csv(filepath)
a=readcsv(filepath)
print(a)
```

```python
a.columns
```

```python
t=a[['prime_genre','user_rating']]
g=t[t.prime_genre=='Games']
ng=t[t.prime_genre!='Games']
print("Mean {0}".format(g.mean()))
print("Mean not games {0}".format(ng.mean()))
```

```python
movies=['free app','Affordable apps','Expensive app','Very Expensive']
rating1=a[a.price==0]
rating2=a[(a.price>0)&(a.price<20)]
rating3=a[(a.price>20)&(a.price<50)]
rating4=a[(a.price>50)]
free=rating1.count()[0]
af_ap=rating2.count()[0]
ea=rating3.count()[0]
ve=rating4.count()[0]
li=[free,af_ap,ea,ve]
plt.pie(li,labels=movies,shadow=True,startangle=90,autopct="%1.1f%%")
plt.show()
```
